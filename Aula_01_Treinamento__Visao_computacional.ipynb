```python
# importing libs
import cv2 as cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import matplotlib.patches as patches
from skimage import io
```

```python
#show image
img = mpimg.imread("cameraman.tif")
plt.imshow(img, cmap='gray'); 
```

```python
print(img)
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
ax.hist(img.ravel(), bins=256)
plt.title("Original Image")
plt.show()
```

```python
histogram , bin_edges = np.histogram(img, bins=256)
```

```python
plt.figure()
plt.title("Hitogram gray")
plt.xlim(0, 255)
plt.plot(bin_edges[0:-1], histogram);
```

```python
#show image with low range pixels [0 - 100]
img_low = plt.imshow(img, cmap='gray', vmin=0, vmax=100)
```

```python
#draw in image
fig, axes = plt.subplots(figsize=(10,7))
axes.imshow(img, cmap='gray');
rect = patches.Rectangle((50,100), 100, 200,linewidth=2, edgecolor='r', facecolor='none' )
axes.set_title("myRoi in Image")
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.add_patch(rect);
```

```python
# Using cv2.rectangle() method ........ BUG
# Draw a rectangle with blue line borders of thickness of 2 px
pointA=(50,100)
pointB = (150, 300)
color=(255, 0, 0)
thickness = 2
image = cv2.rectangle(img, pointA, pointB, color, thickness)
  
# Displaying the image 
plt.imshow(image) # instead cv2.imshow() BUG
```

```python
#img[y, x]
newImg = img[100:300, 50:150];
plt.title("My ROI");
plt.xticks([])
plt.yticks([])
plt.imshow(newImg);
```

```python
#create hist of my Roy
fig, axes = plt.subplots(nrows=1, ncols= 1, figsize=(10,7))
axes.set_title("Histogram of myRoi")
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.hist(newImg.ravel(), bins=256);
```
